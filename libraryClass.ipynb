```python
nota = float(input("Ingrese la nota: "))

if 4 <= nota <= 5:
  print("Buen trabajo")
elif 3 <= nota < 4:
  print("Trabajo aceptable")
elif 0 <= nota < 3:
  print("Debe mejorar")
else:
  print("La nota no es correcta, ingrese una nota nueva")
```

```python
peso = float(input("Ingrese el peso de su paquete en libras "))

if 0 < peso < 1:
  print("La tarifa de envío es de $5")
elif 1 <= peso <= 5:
  print("La tarifa de envío es de $10")
elif 5 < peso <= 10:
  print("La tarifa de envío es de $15")
elif peso > 10:
  print("La tarifa de envío es de $20")
else:
  print("El peso ingresado es incorrecto, ingrese uno nuevo")
```

```python
monto = float(input("Ingrese el monto total de su compra en dolares "))

if monto >= 100:
  print("El descuento que obtiene es de: ", monto * 0.1) #Tambien se puede colocar monto*(1-0.1) y eso me da el valor con el descuento ya aplicado
elif 0 < monto < 100:
  print("El descuento que obtiene es de: ", monto * 0.05)
else:
  print("El monto ingresado es incorrecto, ingrese uno nuevo")
```
