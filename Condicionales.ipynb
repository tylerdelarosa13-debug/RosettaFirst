```python
edad = int(input("Ingrese la edad: "))
boleta = input("Ingrese el tipo de boleta. GENERAL/VIP").upper()

if boleta == "GENERAL":
  if 0 < edad < 12:
    print("El precio es de $8000")
  elif 12 <= edad <= 17:
    print("El precio es de $10000")
  elif 18 <= edad <= 59:
    print("El precio es de $15000")
  elif edad > 59:
    print("El precio es de $7000")
  else:
    print("Rango de edad no válida")
elif:
  if 0 < edad < 12:
    print("El precio es de $12000")
  elif 12 <= edad <= 17:
    print("El precio es de $15000")
  elif 18 <= edad <= 59:
    print("El precio es de $20000")
  elif edad > 59:
    print("El precio es de $12000")
  else:
    print("Edad no válida")
else:
  print("Tipo de boleta no válido")


```

```python
num = int(input("Ingrese el numero "))

if num % 2 == 0 and num % 10 == 0:
  print("El numero ingresado es par y divisible por 10 ")
else:
  print("El numero ingresado no cumple la condicion ")
```

```python
nota = float(input("Ingrese la calificación de sus examenes "))
num = float(input("Ingrese la calificación de sus proyectos "))

if 0 <= nota < 60 or 0 <= num < 70:
  print("No Aprobó")
elif 60 <= nota <= 100 and 70 <= num <= 100:
  print("Aprobó")
else:
  print("Las calificaciones ingresadas no son correctas")
```

```python
edad = int(input("Ingrese la edad "))
licencia = input("¿Tiene licencia vigente? SI/NO ").lower()
alcohol = input("¿Está ebrio? SI/NO ").lower()
por = input("¿Esta con un acompañante? ").lower()

if edad >= 18 and licencia == "si" and alcohol == "no":
  print("Puede conducir")
elif 16 <= edad <= 17 and por == "si" and alcohol == "no":
  print("Puede conducir con acompañante")
else:
  print("No puede conducir")
```
