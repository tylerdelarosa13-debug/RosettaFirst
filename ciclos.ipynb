```python
#Forma mas larga de hacerlo, para eso se usan los ciclos

edad1 = int(input("Ingrese la primera edad "))
edad2 = int(input("Ingrese la segunda edad "))
edad3 = int(input("Ingrese la tercera edad "))
edad4 = int(input("Ingrese la cuarta edad "))
edad5 = int(input("Ingrese la quinta edad "))
edad6 = int(input("Ingrese la sexta edad "))
edad7 = int(input("Ingrese la septima edad "))
edad8 = int(input("Ingrese la octava edad "))
edad9 = int(input("Ingrese la novena edad "))
edad10 = int(input("Ingrese la decima edad "))

promedio = (edad1+edad2+edad3+edad5+edad6+edad7+edad8+edad9+edad10)/10
print("El promedio de edades es: ", promedio)
```

```python
edades = []
n = int(input("Ingrese el numero de edades al que desea calcular el promedio"))

for A in range(n):
  edad = int(input(f"Ingrese la edad {A+1} ")) #print f me deja colocar variables dentro de las comillas en el cual solo me deja colocar caracters que en este caso es {A+1}
  edades.append(edad)
print(f"El promedio de {A+1} edades: ", sum(edades)/len(edades))
```
